Import Package

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Define fftreal

```python
def fftreal(x, y):
    # Combine x and y into a single complex signal
    z = x + 1j * y
    
    # Compute the FFT of the combined signal
    Z = np.fft.fft(z)
    
    # Extract the FFTs of the original signals
    Fx = 0.5 * (Z + np.conj(np.fft.fft(np.conj(z[::-1]))))
    Fy = 0.5j * (Z - np.conj(np.fft.fft(np.conj(z[::-1]))))
    
    return Fx, Fy
```

Example

```python
# Example usage:
x = np.array([1, 2, 3, 4])
y = np.array([4, 3, 2, 1])

Fx, Fy = fftreal(x, y)

print("FFT of x:", Fx)
print("FFT of y:", Fy)
```

Plots

```python
# Plot the results
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot the original signals
axs[0, 0].stem(x)
axs[0, 0].set_title("Original Signal x")
axs[0, 0].set_xlabel("Sample")
axs[0, 0].set_ylabel("Amplitude")

axs[0, 1].stem(y)
axs[0, 1].set_title("Original Signal y")
axs[0, 1].set_xlabel("Sample")
axs[0, 1].set_ylabel("Amplitude")

# Plot the FFTs
axs[1, 0].stem(np.abs(Fx))
axs[1, 0].set_title("FFT of x")
axs[1, 0].set_xlabel("Frequency Bin")
axs[1, 0].set_ylabel("Magnitude")

axs[1, 1].stem(np.abs(Fy))
axs[1, 1].set_title("FFT of y")
axs[1, 1].set_xlabel("Frequency Bin")
axs[1, 1].set_ylabel("Magnitude")

plt.tight_layout()
plt.show()
```
